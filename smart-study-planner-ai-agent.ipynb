```python
class Memory:
    def __init__(self):
        self.store = {}

    def remember(self, key, value):
        self.store[key] = value

    def recall(self, key):
        return self.store.get(key, None)
print("run successfully")
```

```python
class SummarizerAgent:
    def __init__(self, memory):
        self.memory = memory

    def summarize(self, text):
        summary = call_gemini(f"Summarize: {text}")
        self.memory.remember("last_summary", summary)
        return summary
print("run successfully")
```

```python
class QuizAgent:
    def __init__(self, memory):
        self.memory = memory

    def generate_quiz(self, text):
        summary = self.memory.recall("last_summary") or text
        quiz = call_gemini(f"Generate 3 MCQs from: {summary}")
        self.memory.remember("last_quiz", quiz)
        return quiz
print("run successfully")
```

```python
class PlannerAgent:
    def __init__(self, memory):
        self.memory = memory

    def plan_day(self):
        summary = self.memory.recall("last_summary")
        quiz = self.memory.recall("last_quiz")
        plan = f"Today's study plan:\n- Review Summary: {summary}\n- Attempt Quiz: {quiz}"
        self.memory.remember("today_plan", plan)
        return plan
print("run successfully")
```

```python
class Orchestrator:
    def __init__(self):
        self.memory = Memory()
        self.summarizer = SummarizerAgent(self.memory)
        self.quiz_agent = QuizAgent(self.memory)
        self.planner = PlannerAgent(self.memory)

    def run(self, notes_text):
        print("Running Study Planner Agents...\n")
        summary = self.summarizer.summarize(notes_text)
        print("Summary Generated:\n", summary, "\n")

        quiz = self.quiz_agent.generate_quiz(notes_text)
        print("Quiz Generated:\n", quiz, "\n")

        plan = self.planner.plan_day()
        print("Daily Study Plan:\n", plan)
print("run successfully")
```
